```python
#Given three ints, a b c, one of them is small, one is medium and one is large. 
#Print true if the three values are evenly spaced, 
#so the difference between small and medium is the same as the difference between medium and large.

a=int(input())
b=int(input())
c=int(input())
# YOUR CODE HERE
def even_space(a,b,c):
  dummy=[a,b,c]
  d1=[a,b,c]
  d1.remove(min(dummy))
  d1.remove(max(dummy))
  d2=[min(dummy), d1[0],max(dummy)]
  if d2[1]-d2[0]==d2[2]-d2[1]:
    print('evenly spaced')
  else:
    print('not evenly spaced')

  
even_space(a,b,c)        
```

```python
#Write a function to return prime if it is primenumber otherwise return not a prime number. 
def primeCheck(n):
  f=0
  fac=[]
  p=n
  while n>1:
    
    n=n-1
    if p%n==0:
      f=f+1
      fac.append(n)
  if f>1:
    print('its not a prime number')
  else:
    print ('its a prime number')



      
```

```python
#Take two non negative integers from user and pass them to a funntion which will return their GCD.
def factors(n):
  fac=[]
  p=n
  while n>1:
    n=n-1
    if p%n==0:
      
      fac.append(n)
  return fac

def computeGCD(a,b):
  l=set(factors(a))
  m=set(factors(b))
  k=l.intersection(m)
  return max(k)

def findLCM(x,y):
  # choose the greater number
   if x > y:
       greater = x
   else:
       greater = y

   while(True):
       if((greater % x == 0) and (greater % y == 0)):
           lcm = greater
           break
       greater += 1

   return lcm
   
```
